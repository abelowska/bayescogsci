### Exercise 2.5 Conjugate forms 4

```r
# mean = 100; variance = 225
# mean = a/b; variance = a/b^2

a = 10000/225
b = 100/225
```

```r
alpha = a
beta = b
mean = alpha / beta
std = sqrt(alpha / beta^2)

print(mean)
print(std)
```

```r
# observed data
x = c(115,97,79,131)
```

```r
# define data range
range = seq(0, mean + 4*std)

# calculate priors, likelihood and posterior given data x
prior = dgamma(range, alpha, rate = beta)
likelihood = dpois(range, lambda = mean(x))
posterior = dgamma(range, alpha+sum(x), rate = beta+length(x))

# plot results
plot(range, prior, type ="l", ylim=c(0,0.1), col="pink", lwd=3)
lines(likelihood, col="green", lwd=3)
lines(range, posterior,col = "red", lwd=3)

# update posterior with observation x = 200
posterior_2 = dgamma(range, alpha+sum(x)+200, rate = beta+length(x)+1)

# add line to plot
lines(range, posterior_2,col = "orange", lwd=3)

# add a legend
legend(x = "topleft", legend=c("prior", "likelihood", "posterior", "posterior updated"),
       col=c("pink", "green", "red", "orange"),lwd = 2)
```
